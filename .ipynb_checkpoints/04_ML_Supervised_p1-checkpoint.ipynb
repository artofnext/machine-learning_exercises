```python
import math
from random import seed
from random import randint
import numpy as np
from matplotlib import pyplot as plot
from mpl_toolkits.mplot3d import axes3d
import matplotlib.tri as mtri
from matplotlib import cm
```

```python
# Choose the operator OR and declare input & desired vectors
inputV = [[0,0],[1,0],[0,1],[1,1]]
desiredV = [1,-1,-1,-1]
```

```python
# generates set of all possible bit patterns for perceptron with "inputsNumber" inputs
# returns list of bit pattern lists
def inputDataGenerate(inputsNumber = 2):
    result = []
    formatter = "0" + str(inputsNumber) + "b"
    for i in range(2 ** inputsNumber):
        result.append([int(x) for x in format(i, formatter)])
    return result
```

```python
# The "Rectified linear unit" step function for Perceptron
def rectifiedLinearUnit (x):
    if (x > 0): return 1
    else: return -1
```

```python
class Perceptron:
    
    def __init__(self, inputsNumber, alpha = 0.1, wInit = 0.000001):
        self.aplha = alpha
        self.inputsNumber = inputsNumber
        self.isTrained = False
        self.cycles = 0
        self.w = [wInit] * (self.inputsNumber + 1)

    def perceptronFunc (self, xVector):
        result = self.w[0]
        for index, x in enumerate(xVector):
            result += self.w[index + 1] * x
        return rectifiedLinearUnit(result)

    # main traine cycle
    def learn(self, inputV, desiredV):
        dw = [0] * (self.inputsNumber + 1)
        
        for index, x in enumerate(inputV):
            # calculate difference
            difference = desiredV[index] - self.perceptronFunc(x)
            # update terms
            dw[0] += self.aplha * difference
            for index, elemDw in enumerate(dw[1:]):
                dw[index + 1] += x[index] * self.aplha * difference

        for i, elemW in enumerate(self.w):
            self.w[i] += dw[i]
        # if there is no difference, assume Perceptron was trained
        if (sum(dw) == 0.0):
            self.isTrained = True
    
    def traine(self, inputV, desiredV):

        while (not self.isTrained):
            self.learn(inputV, desiredV)
            self.cycles += 1
            # infinite loop preventer
            if (self.cycles > 1000): break
            
    # return outputs based on trained data
    def getOutputV(self, inputV):
        outputV = []
        for x in inputV:
            outputV.append(self.perceptronFunc(x))
        return outputV
 
```

```python
alphaV = []
initWV = []
cyclesV = []

for w in range(0, 10) :
    for a in range(1, 100) :
        alpha = a / 1000000
    
        alphaV.append(alpha)
        initW = w / 100
        initWV.append(initW)
        
        perc2bit = Perceptron(2, alpha, initW)
        perc2bit.traine(inputV, desiredV)
        cyclesV.append(perc2bit.cycles)
```

```python
fig = plot.figure()
ax = plot.axes(projection="3d")
surf = ax.scatter(alphaV, initWV, cyclesV, linewidth=0, antialiased=False)
#fig.colorbar(surf, shrink=0.5, aspect=5)
ax.view_init(elev=40, azim=-60)
ax.set_title('Cycles of traine vs initial values')
ax.set_xlabel('Alpha')
ax.set_ylabel('initial w')
ax.set_zlabel('cycles')
plot.show()
```

This family of curves makes it possible to evaluate the dependence of the computation cycles (iterations through the whole set) on the initial values of w and Alpha (when alpha is 0, the calculations have no sense).
In the case of the OR rule, the best initial value of w is 0. Thus, the process of computing w has one-sided convergence. In the case of large values of w, the approximation process has an oscillatory character, which increases the number of cycles. This is also true for alpha, since the resulting w is proportional to the product of the difference (or error) by the alpha.

2. Try to learn the XOR function

```python
# Choose the operator XOR and declare input & desired vectors
inputV = [[0,0],[1,0],[0,1],[1,1]]
desiredV = [1,-1,-1,1]
```

```python
alphaV = []
initWV = []
cyclesV = []

for w in range(0, 10) :
    for a in range(1, 100) :
        # alpha step
        alpha = a / 1000000
        alphaV.append(alpha)
        # w step
        initW = w / 100
        initWV.append(initW)
        perc2bit = Perceptron(2, alpha, initW)
        perc2bit.traine(inputV, desiredV)
        cyclesV.append(perc2bit.cycles)
```

```python
fig = plot.figure()
ax = plot.axes(projection="3d")
surf = ax.scatter(alphaV, initWV, cyclesV, linewidth=0, antialiased=False)
#fig.colorbar(surf, shrink=0.5, aspect=5)
ax.view_init(elev=40, azim=-60)
ax.set_xlabel('Alpha')
ax.set_ylabel('initial w')
ax.set_zlabel('cycles')
plot.show()
```

As you can see, in the case of the XOR function, the perceptron does not reach the state when the error is zero (infinite loop preventer at 1000 cycles is triggered), that is, the learning process does not converge.
This is because each input has a proportional effect on the output of the perceptron. Each of the inputs has a threshold value for switching the threshold function. For the XOR function, there is no such combination of input values that would satisfy all output states. In other words, the perceptron is a binary classifier with a linear division of the set of input states into two sets. In the case of the OR and AND functions, there is one separation boundary into 2 subsets, which corresponds to the perceptron model. In the case of the XOP function (which is a combination of 2 elementary functions), there are 2 boundaries for dividing subsets into 3, followed by the need to combine two of them. This operation requires a second perceptron.

3. Repeat the process for a bit pattern with 10 inputs and generate the output (-1 or +1) according to a rule. 
Rule: for combinations of inputs that match even numbers, the output is +1, for odd numbers is -1

```python
# helper functions

# rule odd = -1 even = 1
# input list of bit pattern lists
# return list
def desiredOutputOddEven(inputV):
    result = []
    for x in inputV:
        res = 1 if (int("".join(str(elem) for elem in x), 2) % 2) == 0  else -1
        result.append(res)
    return result

# return randomly selected elements as a new list
def getRandomDataset(inputV, percent = .3):
    inputLength = len(inputV)
    outputLength = int(inputLength * percent) + 1
    output = []
    for i in range(outputLength):
        output.append(inputV[randint(0, inputLength - 1)])
    return output

# cumulative error
def calculateError (vector1, vector2):
    error = 0
    for index, v1 in enumerate(vector1):
        error += 0 if (v1 - vector2[index]) == 0 else 1
    return error
```

Let's train the perceptron based on a 30% random sample of all possible options. Let's repeat 30 times for statistics.

```python
# trained with 30% set

# input params
seed(1)
repeats = 30
alpha = 0.1
initW = 0.1
inputsNumber = 10

# output measurements
outputCycles = []
outputErrors = []
learnSucess = []

for i in range(repeats):
    # generate all possible input combinations
    inputV = inputDataGenerate(inputsNumber)
    # randomly select 30%
    inputSetV = getRandomDataset(inputV)
    # generate desired outputs for selected 30%
    desiredV = desiredOutputOddEven(inputSetV)
    # instantiate perceptron instance
    perc3bit = Perceptron(inputsNumber, alpha, initW)
    # traine with randomly select 30% set
    perc3bit.traine(inputSetV, desiredV)
    # save cycles
    outputCycles.append(int(perc3bit.cycles))
    # calculate errors with whole data set 
    outputErrors.append(100 * calculateError(desiredOutputOddEven(inputV), perc3bit.getOutputV(inputV)) / len(inputV))
    # save learn sucess flag
    learnSucess.append(1 if perc3bit.isTrained else 0)
    

#print(outputCycles)
print("Average cycles: " + str(np.mean(outputCycles)))
print("Standard deviation: " + str(np.std(outputCycles)))
#print(learnSucess)
print("Learning success rate: " + str(np.mean(learnSucess) * 100) + "%")
#print(outputErrors)
print("Average output errors: " + str(np.mean(outputErrors)) + "%")
print("Output errors standard deviation: " + str(np.std(outputErrors)))
```

Using 30% of the sample for training gives an average error of 0.016% over the entire dataset. Learning cycles range from 3 to 9 with an average of 5.7.

Let's create a function, that randomly add 1% errors to desired output (it invert output value for certain input combination).


```python
# add randome error to each 100-th output
# rule odd = -1 even = 1
# input list of bit pattern lists
# return list
def desiredOutputRandomError(inputV):
    result = []
    for x in inputV:
        res = 1 if (int("".join(str(elem) for elem in x), 2) % 2) == 0  else -1
        # invert output value with 1% probability
        if (randint(0, 100) == 0): res = 1 if (res == -1) else -1
        result.append(res)
    return result
```

Now, let's train the perceptron based on a 30% random sample of all possible options with this "spoiled" desired output data. Let's repeat 30 times for statistics.

```python
# trained with 30% set and 1% error

seed(1)
repeats = 30
alpha = 0.1
initW = 0.1
inputsNumber = 10

outputCycles = []
outputErrors = []
learnSucess = []

for i in range(repeats):
    # generate all possible input combinations
    inputV = inputDataGenerate(inputsNumber)
    # randomly select 30%
    inputSetV = getRandomDataset(inputV)
    # generate desired outputs for selected 30%
    desiredV = desiredOutputRandomError(inputSetV)

    # instantiate perceptron instance
    perc3bit = Perceptron(inputsNumber, alpha, initW)
    # traine with randomly select 30% set
    perc3bit.traine(inputSetV, desiredV)
    # save cycles
    outputCycles.append(int(perc3bit.cycles))
    # calculate errors with whole data set 
    outputErrors.append(100 * calculateError(desiredOutputOddEven(inputV), perc3bit.getOutputV(inputV)) / len(inputV))
    # save learn sucess flag
    learnSucess.append(1 if perc3bit.isTrained else 0)
    

#print(outputCycles)
print("Average cycles: " + str(np.mean(outputCycles)))
print("Standard deviation: " + str(np.std(outputCycles)))
#print(learnSucess)
print("Learning success rate: " + str(np.mean(learnSucess) * 100) + "%")
#print(outputErrors)
print("Average output errors: " + str(np.mean(outputErrors)) + "%")
print("Output errors standard deviation: " + str(np.std(outputErrors)))
```

As we can see, in 13% of cases, the perceptron does not achieve the convergence of learning, because the data is inconsistent (the execution limit of 1000 cycles is used). Introducing 1% errors significantly degrades the learning outcome.

Now, let's train perceptron on 100% set with 1% spoiled sample to see how it improve traine results.

```python
# trained with 100% set and 1% error

seed(1)
repeats = 30
alpha = 0.1
initW = 0.1
inputsNumber = 10

outputCycles = []
outputErrors = []
learnSucess = []

for i in range(repeats):
    # generate all possible input combinations
    inputV = inputDataGenerate(inputsNumber)
    # randomly select 30%
    #inputSetV = getRandomDataset(inputV)
    # generate desired outputs for 100% set
    desiredV = desiredOutputRandomError(inputV)

    # instantiate perceptron instance
    perc3bit = Perceptron(inputsNumber, alpha, initW)
    # traine with 100% set
    perc3bit.traine(inputV, desiredV)
    # save cycles
    outputCycles.append(int(perc3bit.cycles))
    # calculate errors with whole data set 
    outputErrors.append(100 * calculateError(desiredOutputOddEven(inputV), perc3bit.getOutputV(inputV)) / len(inputV))
    # save learn sucess flag
    learnSucess.append(1 if perc3bit.isTrained else 0)
    

#print(outputCycles)
print("Average cycles: " + str(np.mean(outputCycles)))
print("Standard deviation: " + str(np.std(outputCycles)))
#print(learnSucess)
print("Learning success rate: " + str(np.mean(learnSucess) * 100) + "%")
#print(outputErrors)
print("Average output errors: " + str(np.mean(outputErrors)) + "%")
print("Output errors standard deviation: " + str(np.std(outputErrors)))
```

Although the level of learning convergence has decreased (due to the larger absolute value of the introduced errors), the level of errors of the trained perceptron is lower. Conclusion: in the case of erroneous data, increasing the training sample allows you to achieve a better result.
